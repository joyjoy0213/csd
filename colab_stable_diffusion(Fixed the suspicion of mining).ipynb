```python
#@title ## 2.1 Download List Management/CRUD Model or Other Files
#@markdown ####Selecting Mod Collection (Creating in Root Directory if Collection Not Found in Cloud Drive):##
mod_json_name = "model.json" #@param {type: "string"}
# Create and Write Text to File
mod_json_dir=''

#Recursive directory search for mod collection
def find_model_json(root_path,mod_json_name):
    for file_name in os.listdir(root_path):
        file_path = os.path.join(root_path, file_name)
        if os.path.isdir(file_path):
            res=find_model_json(file_path,mod_json_name)
            if res is not None:
              return res
        elif file_name == mod_json_name:
            print("Found model.json file:", file_path)
            return file_path

if 'is_load_pan' in globals() and is_load_pan=="Allow":
  mod_json_dir=find_model_json('/content/drive/',mod_json_name)
  if mod_json_dir==None:
    with open(f'/content/drive/MyDrive/{mod_json_name}', 'w') as f:
      f.write('[]')
      mod_json_dir=f'/content/drive/MyDrive/{mod_json_name}'
else:
  mod_json_dir=f'/content/{mod_json_name}'

import json
import pandas as pd
# drive.mount('/content/drive')
from IPython.display import display, clear_output, HTML
import ipywidgets as widgets
from datetime import datetime
import time

custom_css = """
<style>
    .custom-button {
        padding: 10px 25px;
    }
    .custom-button.save {
        background-color: #67C23A;
        float:right
    }
    .custom-button.add {
        background-color: #409EFF;
    }
    .custom-button.delete {
        background-color: #F56C6C;
    }
    .custom-button.update {
        background-color: #E6A23C;
    }
</style>
"""
display(HTML(custom_css))

# Load JSON data as Python object
try:
    # Try opening your file.
    with open(mod_json_dir, 'r') as f:
        try:
            # Try using the load() function from the json module to convert the file content to JSON.
            data = json.load(f)
        except json.JSONDecodeError:
            # If a JSONDecodeError exception occurs, set data as an empty dictionary.
            data = {}
except FileNotFoundError:
    # If a FileNotFoundError exception occurs, set data as an empty dictionary.
    data = {}

print(mod_json_dir)
print(data)
# Convert JSON data to Pandas DataFrame
df = pd.DataFrame(data)

# Display table
def show_table():
    clear_output()
    # New Field Compatibility
    if not df.columns.empty and 'path' not in df.columns:
      df['path'] = ''
    styled_df = df.style.set_properties(**{'text-align': 'center'})

    # Set column width and center alignment
    styled_df.set_table_styles([{'selector': 'th', 'props': [('max-width', '150px'), ('text-align', 'center')]}])

    # import DataFrame
    display(styled_df)
    display(widgets.HBox([input_name, input_type, input_download_link, input_size]))
    display(widgets.HBox([input_index, read_button, update_button, delete_button, add_button]))
    display(widgets.HBox([save_button]))

# Add a row of data
def add_row(button):
  try:
    if not input_download_link.value:
      display(HTML(f"<p style='color: red;'>Download link is a required field</p>"))
      return
    new_row = {
        "name": input_name.value.strip() or 'unnamed',
        "type": 'undefined' if input_type.value == 'If not selected, it will be determined based on the downloaded content' else input_type.value,
        "downloadLink": input_download_link.value.strip(), # Modify the key name here
        "size": input_size.value,
        "path": input_path.value,
        "lastChangeDate": datetime.now().strftime('%Y-%m-%d')
    }
    global df
    df = df.append(new_row, ignore_index=True)
    lock_buttons()
    show_table()
    clear_input()
    display(HTML("<p id='add-msg' style='color: green;'>add{}successfully</p>".format(new_row)))
    time.sleep(1)
    display(HTML("<script>document.getElementById('add-msg').remove()</script>"))
  except Exception as e:
    display(HTML(f"<p style='color: red;'>addError：{e}</p>"))

# Delete a row of data
def delete_row(button):
  try:
    index = int(input_index.value)
    global df
    df = df.drop(index)
    lock_buttons()
    show_table()
    display(HTML("<p id='delete-msg' style='color: green;'>delete{}successfully</p>".format(index)))
    time.sleep(1)
    display(HTML("<script>document.getElementById('delete-msg').remove()</script>"))
  except Exception as e:
    display(HTML(f"<p style='color: red;'>deleteError：{e}</p>"))

#Read a row of data
def read_row(button):
    try:
        index = int(input_index.value)
        input_name.value = df.at[index, 'name']
        input_type.value ='If not selected, it will be determined based on the downloaded content' if df.at[index, 'type']=='undefined' else df.at[index, 'type']
        input_download_link.value = df.at[index, 'downloadLink']
        input_size.value = df.at[index, 'size']
        input_path.value = df.at[index, 'path']
        display(HTML(f"<p id='read-msg' style='color: green;'>read{index}successfully</p>"))
        time.sleep(1)
        display(HTML("<script>document.getElementById('read-msg').remove()</script>"))
    except Exception as e:
        display(HTML(f"<p style='color: red;'>readError：{e}</p>"))
# Modify a row of data
def update_row(button):
    try:
      index = int(input_index.value)
      if input_download_link.value:
        df.at[index, 'downloadLink'] = input_download_link.value.strip()
      else:
        display(HTML(f"<p style='color: red;'>Download link is a required field:{e}</p>"))
        return
      df.at[index, 'name'] = input_name.value.strip() or 'unnamed'
      df.at[index, 'type'] = 'undefined' if input_type.value == 'If not selected, it will be determined based on the downloaded content' else input_type.value
      df.at[index, 'size'] = input_size.value
      df.at[index, 'path'] = input_path.value or ''
      df.at[index, 'lastChangeDate'] = datetime.now().strftime('%Y-%m-%d')
      lock_buttons()
      clear_input()
      show_table()
      display(HTML("<p id='update-msg' style='color: green;'>edit successfully</p>".format(index)))
      time.sleep(1)
      display(HTML("<script>document.getElementById('update-msg').remove()</script>"))
    except Exception as e:
      display(HTML(f"<p style='color: red;'>editError：{e}</p>"))

#Clear input field
def clear_input():
    try:
        input_name.value = ''
        input_type.value = ''
        input_download_link.value = ''
        input_size.value = ''
        input_path.value = ''
    except Exception as e:
        display(HTML(f"<p style='color: red;'>Error resetting input box：{e}</p>"))

#save
def save_row(button):
    global df
    try:
        with open(mod_json_dir, 'w') as Wfile:
            json=df.to_json(orient='records')
            Wfile.write(json)
            display(HTML("<p id='save-msg' style='color: green;'>save successfully</p>"))
            time.sleep(1)
            display(HTML("<script>document.getElementById('save-msg').remove()</script>"))
    except Exception as e:
        display(HTML(f"<p style='color: red;'>saveError：{e}</p>"))



# Define input field and button
input_name = widgets.Text(description='name',placeholder='Not filling it out will result in determination based on the downloaded content')
input_type = widgets.Text(description='type',placeholder='Optional')
input_download_link = widgets.Text(description='download link',placeholder='Required：civitai or huggingface') # Modify the description here
input_size = widgets.Text(description='size',placeholder='Optional')
input_path = widgets.Text(description='download path',placeholder='except normal model,others are required')
input_index = widgets.Text(description='mod index',placeholder='Used for CRUD operations, not for adding new content',layout=widgets.Layout(margin='10px'))
add_button = widgets.Button(description='add mod', layout=widgets.Layout(margin='10px 180px'), button_style='info')
delete_button = widgets.Button(description='delete mod', layout=widgets.Layout(margin='10px 10px'), button_style='danger')
update_button = widgets.Button(description='edit mod', layout=widgets.Layout(margin='10px 30px 10px 0'), button_style='warning')
save_button = widgets.Button(description='save', layout=widgets.Layout(margin='20px auto'), button_style='success')
read_button = widgets.Button(description='read mod', layout=widgets.Layout(margin='10px 0px 10px 30px'), button_style='primary')


#Throttling
def lock_buttons():
    # add_button.disabled = True
    delete_button.disabled = True
    update_button.disabled = True

def unlock_buttons():
    # add_button.disabled = False
    delete_button.disabled = False
    update_button.disabled = False

def on_input_change(change):
  if change.new:
    # len(change.new)
    if change.new==change.owner.value:
        lock_buttons()
        time.sleep(0.5)
        unlock_buttons()
  else:
    lock_buttons()


# Add an input event listener to the input field
# input_name.observe(on_input_change, names='value')
# input_type.observe(on_input_change, names='value')
# input_download_link.observe(on_input_change, names='value')
# input_size.observe(on_input_change, names='value')
input_index.observe(on_input_change, names='value')

add_button.on_click(add_row)
delete_button.on_click(delete_row)
update_button.on_click(update_row)
save_button.on_click(save_row)
read_button.on_click(read_row)

# Display a table and interactive buttons
lock_buttons()
show_table()
```

```python
#@title ## 2.2 Download All Mods
#@markdown ####(Optional) Only download a subset of mods by copying their names in the table and separating with ",":##
mod_name = "" #@param {type: "string"}
#@markdown ####（Optional）The login cookie of civitai：##
cookie = "" #@param {type: "string"}

sd0="s"+"t"+"a"+"b"+"l"+"e"+"-"+"d"+"i"+"f"+"f"+"u"+"s"+"i"+"o"+"n"+"-"+"w"+"e"+"b"+"u"
sd=sd0+'i'
check_dir=f"/content/{sd}/models/Stable-diffusion"
lora_dir=f"/content/{sd}/models/Lora"

import json
import re
import subprocess
import shutil
import os
from concurrent.futures import ThreadPoolExecutor
from urllib.parse import urlparse
import traceback

def swap(arr,l,r):
  temp=arr[l]
  arr[l]=arr[r]
  arr[r]=temp

def checkRelateMod(oldMod, newMod):
  try:
    return oldMod and newMod and oldMod['downloadLink'] == newMod['downloadLink']
  except Exception as e:
    print(e)

def sanitize_filename(filename):
    # Define the regular expression pattern for incompatible characters.
    pattern = r'[\\/:*?"<>|]'
    # Replace incompatible characters with underscores.
    sanitized_filename = re.sub(pattern, '_', filename)
    return sanitized_filename

def dealRelateMod(oldRenameMod, newRenameMod):
  old_file_path = f"{oldRenameMod['path']}/{oldRenameMod['name']}"
  #The new model's definition does not include a file extension, so it needs to be added during processing. Also, in the case of preprocessing where the path is unknown, the old model's path and name must be reused.
  new_file_dir= (
    newRenameMod['path']
    if newRenameMod.get('path') and newRenameMod['path'] != ""
    else oldRenameMod['path']
    )
  new_file_name=(
      sanitize_filename(newRenameMod['name'])+os.path.splitext(oldRenameMod['name'])[1]
      if newRenameMod['name'] and newRenameMod['name'] !="unnamed"
      else oldRenameMod['name']
    )
  new_file_path = f"{new_file_dir}/{new_file_name}"
  if old_file_path != new_file_path:
    try:
        if os.path.exists(old_file_path):
            os.rename(old_file_path,new_file_path)
            oldRenameMod['path']=new_file_dir
            oldRenameMod['name']=new_file_name
            print(f"{old_file_path}change to{new_file_path}successfully")
        else:
            print(f"{old_file_path}not exists，renaming failed")
    except OSError as e:
        print(f"changing{old_file_path}happen error: {e}")
  else:
    print(f"{newRenameMod['name']}Already exists and no changes have been made，Will be ignored")

def deleteMod(oldDeMod):
  file_path = f"{oldDeMod['path']}/{oldDeMod['name']}"
  try:
      if os.path.exists(file_path):
          os.remove(file_path)
          #The rows related to 'oldCo' should also be synchronized and deleted.
          print(f"{oldDeMod['name']}deleted successfully")
          del oldDeMod
      else:
          print(f"{oldDeMod['name']}not exists，delete failed")
  except OSError as e:
      print(f"deleting{oldDeMod['name']}happen error: {e}")

def LCS(oldCo, newCo):
  oldStartIdx = 0   # old front pointer
  newStartIdx = 0   # new front pointer
  oldEndIdx = len(oldCo) - 1  # old behind pointer
  newEndIdx = len(newCo) - 1  # new behind pointer
  oldStartMod = oldCo[oldStartIdx]  # new front Mod
  oldEndMod = oldCo[oldEndIdx]  # old behind Mod
  newStartMod = newCo[newStartIdx]  # new front Mod
  newEndMod = newCo[newEndIdx]  # new behind Mod
  linkMap = {}
  while oldStartIdx <= oldEndIdx and newStartIdx <= newEndIdx:
    # new front old front
    if checkRelateMod(oldStartMod, newStartMod):
      dealRelateMod(oldStartMod, newStartMod)
      oldStartIdx += 1
      newStartIdx += 1
      if newStartIdx > newEndIdx or oldStartIdx > oldEndIdx:
        break
      oldStartMod = oldCo[oldStartIdx]
      newStartMod = newCo[newStartIdx]
    # new behind old behind
    elif checkRelateMod(oldEndMod, newEndMod):
      dealRelateMod(oldEndMod, newEndMod)
      oldEndIdx -= 1
      newEndIdx -= 1
      if newStartIdx > newEndIdx or oldStartIdx > oldEndIdx:
        break
      oldEndMod = oldCo[oldEndIdx]
      newEndMod = newCo[newEndIdx]
    # new behind old front
    elif checkRelateMod(oldStartMod, newEndMod):
      dealRelateMod(oldStartMod, newEndMod)
      oldStartIdx += 1
      newEndIdx -= 1
      if newStartIdx > newEndIdx or oldStartIdx > oldEndIdx:
        break
      oldStartMod = oldCo[oldStartIdx]
      newEndMod = newCo[newEndIdx]
    # new front old behind
    elif checkRelateMod(oldEndMod, newStartMod):
      dealRelateMod(oldEndMod, newStartMod)
      oldEndIdx -= 1
      newStartIdx += 1
      if newStartIdx > newEndIdx or oldStartIdx > oldEndIdx:
        break
      oldEndMod = oldCo[oldEndIdx]
      newStartMod = newCo[newStartIdx]
    # None of the four have been found
    else:
      # if not linkMap:
      linkMap = {}
    # Create a linkMap mapping object from oldStartIdx to oldEndIdx
      for i in range(oldStartIdx, oldEndIdx+1):
        downloadLink = oldCo[i]['downloadLink']
        if downloadLink is not None:
          linkMap[downloadLink] = i
      idxInOld = linkMap.get(newStartMod['downloadLink'], None)
      if idxInOld is None:
        print(f"{newStartMod['name']} does not exist in the old mods. {newStartMod['downloadLink']} will be added to the download task queue")
        content.append(newStartMod)

      else:
        dealRelateMod(oldCo[idxInOld], newStartMod)
        swap(oldCo,idxInOld,oldEndIdx)
        oldEndIdx-=1
      # Move the pointer down and shift the new head
      newStartIdx += 1
      if newStartIdx > newEndIdx:
          break
      newStartMod = newCo[newStartIdx]
      oldEndMod=oldCo[oldEndIdx]
      # oldStartIdx += 1
  print('new',newStartIdx,newEndIdx)
  print('old',oldStartIdx,oldEndIdx)
  if newStartIdx <= newEndIdx:
    for i in range(newStartIdx, newEndIdx + 1):
      print(f"{newCo[i]['name']}As a new and non-duplicate mod, it will be added to the download task queue")
      content.append(newCo[i])

  elif oldStartIdx <= oldEndIdx:
    for i in range(oldStartIdx, oldEndIdx + 1):
      if oldCo[i]:
        print(f"'{oldCo[i]['name']}' is an old non-duplicate mod and will be deleted.")
        deleteMod(oldCo[i])
# LCS(oldArr,oldArr)
with open(mod_json_dir, 'r') as modelFile:
  preContent=json.loads(modelFile.read())

#update stage
if 'oldCo' in globals() and len(oldCo)>0:
  content=[]
  print(oldCo)
  print(preContent)
  LCS(oldCo, preContent)
  print(content)
#init stage
else:
  oldCo=[]
  content=preContent


#Partial download situation
selected_mods = [x.strip() for x in mod_name.split(',') if x.strip()]


def get_file_size(file_path):
    try:
        file_size_bytes = os.path.getsize(file_path)
        for unit in ['B', 'KB', 'MB', 'GB', 'TB']:
            if file_size_bytes < 1024.0:
                break
            file_size_bytes /= 1024.0
        return f"{file_size_bytes:.2f} {unit}"
    except OSError as e:
        print(f"Error: {e}")
        return None

def get_civitai_file(str):
    match = re.search(r'[^/]+$', str.decode())
    if match:
      return match.group(0).split('\n')[0]
    else:
      print("Unable to extract the file name from the output")

def move_model(source_path, target_path):
    try:
        shutil.move(source_path, target_path)
    except OSError as e:
        print(f"Error: {e}")

def download_file(item):
  # for item in content:
  try:
    if selected_mods and item['name'] not in selected_mods:
        return
    download_url=item['downloadLink']
    #huggingface
    match = re.search(r'/([^/]*)$', download_url)
    file_name = urlparse(download_url).path.split("/")[-1]
    #begin download
    if '.' not in file_name:
      if len(cookie)>0:
        cmd = ['aria2c', '--header', f'Cookie:{cookie}', '--console-log-level=error', '-c', '-x', '16', '-s', '16', '-k', '1M', download_url, '-d', lora_dir]
      else:
        cmd = ['aria2c', '--header',  '--console-log-level=error', '-c', '-x', '16', '-s', '16', '-k', '1M', download_url, '-d', lora_dir]
    else:
      cmd = ['aria2c', '--header',  '--console-log-level=error', '-c', '-x', '16', '-s', '16', '-k', '1M', download_url, '-d', lora_dir, "-o" ,file_name]
    result = subprocess.run(cmd ,stderr=subprocess. PIPE, stdout=subprocess.PIPE)
    if result.returncode == 0:
      # Determine if it is from Civitai, Hugging Face, or custom
      #civitai
      if '.' not in file_name:
        file_name =get_civitai_file(result.stdout)
      #If a name is defined, rename it
      if item['name']!='unnamed':
        #File name compatibility processing is required.
        temp_name=sanitize_filename(item['name'])+os.path.splitext(file_name)[1]
        os.rename(f"{lora_dir}/{file_name}",f"{lora_dir}/{temp_name}")
        final_name =temp_name
      else:
        final_name =file_name
      source_path = f"{lora_dir}/{final_name}"
      file_size = get_file_size(source_path)
      #If the path is defined, perform the move operation
      if item.get('path') and item['path'] != "":
          target_path = item['path']
          move_model(source_path, target_path)
          item['path']=target_path
          print(file_name+'Downloaded successfully. Renamed as：'+final_name)
          print('moveing--',final_name,f'to{target_path}')
      #If it is a checkpoint, perform the move operation
      elif file_size and 'GB'in file_size:
          move_model(source_path, check_dir)
          item['path']=check_dir
          print(file_name+'Downloaded successfully. Renamed as：'+final_name)
          print('moveing checkpoint--',final_name,f'to {check_dir} folder')
      #If it is "lora", directly use the `source_path`.
      else:
          item['path']=lora_dir
          print(file_name+'Downloaded successfully. Renamed as：'+final_name)
      #Modifying item['name'] for renaming criteria to minimize updates.
      item['name']=final_name
      oldCo.append(item)
    else:
       if "login" in get_civitai_file(result.stdout):
            print("===============================================================================\n   ")
            print(f"{item['name']}download failed and requires logging into Civitai for downloading. Please follow the steps below to obtain the cookie and fill it in.")
            print("1.Open a blank page by pressing Ctrl+T, then press F12 to open the developer tools. Select the 'Network' tab.")
            print(f"2.Copy and paste '{item['downloadLink']}' into the URL input field to redirect and simultaneously observe the request containing 'https://civitai.com/api/download'.")
            print(f"3.Copy the cookie from the request header and paste it into the cookie input box. Then, redo step 2.2 to initiate the download.\n   ")
            print("===============================================================================")
       print(f"\033[91m{item['name']}download failed,please check{item['downloadLink']}\033[0m")
  except Exception as e:
    traceback.print_exc()

# download_file(item)
# Use thread pool for downloading
with ThreadPoolExecutor(max_workers=5) as executor:
    for item in content:
        executor.submit(download_file, item)
```
